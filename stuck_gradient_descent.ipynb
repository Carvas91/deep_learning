```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def f(x):
    return np.cos(2 * np.pi * x) + x**2

def deriv(x):
    return -2 * np.pi * np.sin(2 * np.pi * x) + 2 * x

x = np.linspace(-2, 2, 500)
local_min = 0  # Bad initialization!

lr = 0.01 #Learning rate
epochs = 100

for _ in range(epochs):
    grad = deriv(local_min)
    local_min = local_min - lr * grad

# Plot the function and its derivative
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='y', linewidth=2)
plt.plot(x, deriv(x), label='dy', linewidth=2, color='orange')

# Red dots on both function and derivative at x = 0
plt.plot(local_min, f(local_min), 'ro')
plt.plot(local_min, deriv(local_min), 'ro')

# Annotation pointing to the stuck point
plt.annotate('Stuck here: x = 0\nslope = 0, not a minimum',
             xy=(local_min, f(local_min)),       # point to annotate
             xytext=(-1.5, 5),                   # location of the annotation text
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=12,
             color='black')

# Labels and title with LaTeX
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r"$f(x) = \cos(2\pi x) + x^2$  and  $f'(x) = -2\pi \sin(2\pi x) + 2x$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
