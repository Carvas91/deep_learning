```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def f(x):
    return np.cos(2 * np.pi * x) + x**2

def deriv(x):
    return -2 * np.pi * np.sin(2 * np.pi * x) + 2 * x

x = np.linspace(-2, 2, 500)
local_min = 0  # Bad initialization!

lr = 0.01 #Learning rate
epochs = 100

for _ in range(epochs):
    grad = deriv(local_min)
    local_min = local_min - lr * grad

# Plot
plt.plot(x, f(x), label='y')
plt.plot(x, deriv(x), label='dy')
plt.plot(local_min, f(local_min), 'ro')      # Red dot on y
plt.plot(local_min, deriv(local_min), 'ro')  # Red dot on dy
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title("Gradient Descent stuck at x = 0")
plt.show()
```
